```python
api_wrapper=WikipediaAPIWrapper(top_k_results=1,doc_content_chars_max=200)
wiki=WikipediaQueryRun(api_wrapper=api_wrapper)
```

```python
!pip install -q langchain langchain-community faiss-cpu transformers torch numpy

from langchain_community.document_loaders import WebBaseLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter

# ✅ Load Web Content
loader = WebBaseLoader("https://docs.smith.langchain.com/")
docs = loader.load()

# ✅ Split Documents for Better Chunking
documents = RecursiveCharacterTextSplitter(
    chunk_size=1000, chunk_overlap=200
).split_documents(docs)

# ✅ Use Open-Source Embeddings (Replacing OpenAIEmbeddings)
embedding_model = HuggingFaceEmbeddings(model_name="BAAI/bge-m3")

# ✅ Store in FAISS Vector Database
vectordb = FAISS.from_documents(documents, embedding_model)

# ✅ Create Retriever
retriever = vectordb.as_retriever()

print("✅ Retriever created successfully!")
```

```python
!pip install -q langchain langchain-community transformers torch numpy

from langchain_community.llms import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# ✅ Choose an Open-Source LLM (Mistral or Falcon)
model_name = "tiiuae/falcon-7b-instruct"  # Alternative: "tiiuae/falcon-7b-instruct"

# ✅ Load Tokenizer & Model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype="auto", device_map="auto")

# ✅ Create a Text-Generation Pipeline
text_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=200,
    pad_token_id=tokenizer.eos_token_id  # ✅ Fixes padding token warning
)

# ✅ Wrap the LLM in LangChain
llm = HuggingFacePipeline(pipeline=text_pipeline)
```

```python
from langchain.agents import initialize_agent, AgentType, AgentExecutor
from langchain.tools import Tool
```

```python
# ✅ Define Example Tools
def search_tool(query):
    return f"Searching for: {query}"  # Replace with an actual search API if needed

tools = [
    Tool(name="Search", func=search_tool, description="Use this tool to search for information")
]

# ✅ Initialize Agent (Replaces `create_openai_tools_agent`)
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)
```

```python
# ✅ Run the Agent (Fixed Invocation)
response = agent.invoke("How can AI help in healthcare?")

print("✅ AI Response:", response)
```
